# Dealing with Samsung

```python
df = samsung.copy()
df.head()
```

```python
df.describe()
```

## How does Android Version size affect price 

```python
sns.regplot(data = df, x = 'android_version', y = 'price')
```

* This tells us that the android version of samsung phones contribute slightly to an increase in prices of samsung phones

## How does the ram size affect price

```python
sns.regplot(data = df, x = 'ram', y = 'price')
```

* Here we can see that the amount of ram plays a huge role in the price of samsung phones

> A huge increase in ram leads to a hugge increase in price

## How does the battery power affect the price

```python
sns.regplot(data = df, x = 'battery', y = 'price')
```
